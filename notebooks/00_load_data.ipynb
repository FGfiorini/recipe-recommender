```python
import pandas as pd
import numpy as np

# Definimos el rango de 2 meses
fecha_inicio = pd.to_datetime('2026-05-01')
fecha_fin = pd.to_datetime('2026-06-30')

# Calculamos la cantidad de días en el intervalo
dias_rango = (fecha_fin - fecha_inicio).days

# Generamos un offset aleatorio de días para cada una de las 1090 filas
# np.random.randint generará valores entre 0 y 60
offsets_aleatorios = np.random.randint(0, dias_rango + 1, size=len(df))

# Creamos la columna de fecha
df['date'] = fecha_inicio + pd.to_timedelta(offsets_aleatorios, unit='D')

# Ordenamos por fecha para mantener la estructura de serie de tiempo
df = df.sort_values('date').reset_index(drop=True)

# Verificamos la distribución
print(df['date'].value_counts().head()) # Ver cuántas recetas cayeron por día
```

```python
df.head()
```

```python
# Eliminamos la columna 'Unnamed: 0' si existe (a veces aparece al guardar CSVs)
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
# Guardamos el nuevo CSV con la columna de fecha
df.to_csv('recetas.csv', index=False)
```
